```python
# Here in Ticket_Price_Cr Feature can able to find that the values are in inproper formate
# Case_1: ₹100.0 Cr 
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].astype(str).str.replace('₹','').str.replace('Cr','').str.strip()
```

```python
df['Ticket_Price_Cr'].value_counts()
```

```python
# Case_2:
# 14.649171964451835 are in high precision values so we are converting to two decimals
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].round(2)
```

```python
# case_3: Ticket_Price_Cr are in object D-type so we are converting to numeric
df['Ticket_Price_Cr'] = pd.to_numeric(df['Ticket_Price_Cr'],errors = 'coerce')
```

```python
# we are going to find over all null values and Percentage of missing values in df
column_list = list(df.columns)
print(column_list)
for i in column_list :
    total_number_of_null_values = df[i].isnull().sum()
    print(f'Total number of null values for {i}',total_number_of_null_values)
    Total_percentage_of_missing_values = (total_number_of_null_values / df.shape[0]) * 100
    print(f'Total Percentage of null values for {i}',Total_percentage_of_missing_values)
```

```python
# Total Percentage of null values for Ticket_Price_Cr 9.913
# The Null_value percentage is greater than > 5 % , if we drop out it will lead to Data Loss 
# So we are now handling the missing values and outlies detection
import matplotlib.pyplot as plt
plt.hist(df['Ticket_Price_Cr'],bins = 30,color = 'skyblue',edgecolor = 'black')
plt.title('Histogram for Ticket_Price_Cr')
plt.xlabel('Ticket Price')
plt.ylabel ('Frequency')
plt.show()
```

```python
# Now we are filing null values in Ticket_Price_Cr 
# By not only applying median to those missing values , we are seperating the BHK and performing for meaning full analysis
median_val = df.loc[df['Configuration'].isin(['5Bhk+ ', '5BHK+', '5bhk+']),'Ticket_Price_Cr'].median()
median_val
df.loc[df['Ticket_Price_Cr'].isnull() & df['Configuration'].isin(['5Bhk+', '5BHK+', '5bhk+']),'Ticket_Price_Cr'] = median_val
median_val_1 = df.loc[df['Configuration'].isin(['4BHK', '4Bhk', '4bhk']),'Ticket_Price_Cr'].median()
median_val_1
df.loc[df['Ticket_Price_Cr'].isnull() & df['Configuration'].isin(['4BHK', '4Bhk', '4bhk']),'Ticket_Price_Cr'] = median_val_1
median_val_2 = df.loc[df['Configuration'].isin(['3BHK ', '3bhk', '3Bhk']),'Ticket_Price_Cr'].median()
median_val_2
df.loc[df['Ticket_Price_Cr'].isnull() & df['Configuration'].isin(['3BHK ', '3bhk', '3Bhk']),'Ticket_Price_Cr'] = median_val_2
```

```python
# By analysing the Unit_Size_Sqft can able to find that -1.0 as 500 Entries so data is invalid we need to consider as NaN
# Replacing (-1.0) invalid as Nan
df['Unit_Size_Sqft'] = df['Unit_Size_Sqft'].replace(-1.0,np.nan)
```

```python
# Next DataBase insertion
! pip install mysql-connector-python


import mysql.connector


mydb = mysql.connector.connect(
        host = "localhost",
        user = "root",
        password = "12345",
        database = "luxury_housing_sales_analysis",
        autocommit = True)
mycursor = mydb.cursor()

mycursor.execute("""
CREATE TABLE IF NOT EXISTS housing_data (
    Property_ID VARCHAR(255), Micro_Market VARCHAR(255), Project_Name VARCHAR(255), Developer_Name VARCHAR(255),
       Unit_Size_Sqft FLOAT, Configuration VARCHAR(255), Ticket_Price_Cr FLOAT,
       Transaction_Type VARCHAR(255), Buyer_Type VARCHAR(255), Purchase_Quarter DATE,
       Connectivity_Score FLOAT, Amenity_Score FLOAT, Possession_Status VARCHAR(255),
       Sales_Channel VARCHAR(255), NRI_Buyer VARCHAR(255), Locality_Infra_Score FLOAT,
       Avg_Traffic_Time_Min INT, Buyer_Comments TEXT
)
""")


for _, row in df.iterrows():
    sql = """INSERT INTO housing_data  
            (`Property_ID`, `Micro_Market`, `Project_Name`, `Developer_Name`,
            `Unit_Size_Sqft`, `Configuration`, `Ticket_Price_Cr`,
            `Transaction_Type`, `Buyer_Type`, `Purchase_Quarter`,
            `Connectivity_Score`, `Amenity_Score`, `Possession_Status`,
            `Sales_Channel`, `NRI_Buyer`, `Locality_Infra_Score`,
            `Avg_Traffic_Time_Min`, `Buyer_Comments`)
            VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
            """
    mycursor.execute(sql,tuple(row))
mydb.commit()
mycursor.close()
mydb.close()
```

```python
# Market Trends
# How have luxury housing bookings changed quarter by quarter across micro-markets?
mycursor.execute('SELECT Purchase_Quarter, Micro_Market, COUNT(Property_ID) AS Total_Bookings FROM housing_data GROUP BY Purchase_Quarter, Micro_Market ORDER BY Purchase_Quarter, Micro_Market;')
Data = mycursor.fetchall()
df_1 = pd.DataFrame(Data,columns=['Purchase_Quarter', 'Micro_Market','Total_Bookings'])
df_1.to_csv('df_1',index= False)
```

```python
# Builder Performance
# Which builders have the highest total ticket sales and how do they rank in terms of average ticket size?
mycursor.execute('select Developer_Name ,sum(Ticket_Price_Cr) as Total_sum, avg(Ticket_Price_Cr) as avg_Ticket_price ,Rank() over (order by avg(Ticket_Price_Cr) DESC) as Price_Rank from housing_data group by Developer_Name order by avg_Ticket_price DESC;')
Data = mycursor.fetchall()
df_2 = pd.DataFrame(Data,columns=['Developer_Name', 'Total_sum_Ticket_price','avg_Ticket_price','Rank'])
df_2.to_csv('df_2',index= False)
```

```python
# Is there a correlation between amenity score and booking success rate?
mycursor.execute('select avg(Amenity_Score) as avg_amenity_score,Micro_Market,count(Property_ID) as total_bookings from housing_data group by Micro_Market order by Micro_Market ;')
Data = mycursor.fetchall()
df_3 = pd.DataFrame(Data,columns=['avg_amenity_score','Micro_Market','total_bookings'])
df_3.to_csv('df_3',index = False)
```

```python
# Which micro-markets have the highest and lowest booking conversion rates?
mycursor.execute('select Micro_Market, count(Project_Name) as total_count_of_projects from housing_data group by Micro_Market order by Micro_Market;')
Data = mycursor.fetchall()
df_4 = pd.DataFrame(Data,columns=['Micro_Market','total_count_of_projects'])
df_4.to_csv('df_4',index = False)
```

```python
# What are the most in-demand housing configurations (e.g., 3BHK, 4BHK)?
mycursor.execute('select count(Property_ID) as Total_booking_count,Configuration from housing_data group by Configuration ;')
Data = mycursor.fetchall()
df_5 = pd.DataFrame(Data,columns=['Total_booking_count','Configuration'])
df_5.to_csv('df_5',index = False)
```

```python
# Which sales channels contribute most to successful bookings?
mycursor.execute('select Sales_Channel, count(Property_ID) as Total_Property_Count from housing_data group by Sales_Channel order by Sales_Channel;;')
Data = mycursor.fetchall()
df_6 = pd.DataFrame(Data,columns=['Sales_Channel','Total_Property_Count'])
df_6.to_csv('df_6',index = False)
```

```python
# Which builders dominate the market each quarter?
mycursor.execute('select sum(Ticket_Price_Cr) as Total_price ,Purchase_Quarter,Developer_Name from housing_data group by Developer_Name,Purchase_Quarter;')
Data = mycursor.fetchall()
df_7 = pd.DataFrame(Data,columns=['Total_price','Purchase_Quarter','Developer_Name'])
df_7.to_csv('df_7',index = False)
```

```python
#  How does possession status affect buyer type and booking decisions?
mycursor.execute('SELECT Possession_Status,Buyer_Type,Transaction_Type, COUNT(Property_ID) AS Total_Count FROM housing_data GROUP BY Possession_Status, Buyer_Type, Transaction_Type ORDER BY Possession_Status, Buyer_Type;')
Data = mycursor.fetchall()
df_8 = pd.DataFrame(Data,columns=['Possession_Status','Buyer_Type','Transaction_Type','Total_Count'])
df_8.to_csv('df_8',index = False)
```

```python
# Where are most luxury housing projects concentrated within Bangalore?
mycursor.execute(' select Ticket_Price_Cr , Micro_Market from housing_data group by Micro_Market,Ticket_Price_Cr order by Ticket_Price_Cr DESC LIMIT 15;')
Data = mycursor.fetchall()
df_9 = pd.DataFrame(Data,columns=['Ticket_Price_Cr','Micro_Market'])
df_9.to_csv('df_9',index = False)
```

```python
# Who are the top 5 builders in terms of revenue and booking success?
mycursor.execute(' select count(Property_ID) as Total_bookings, sum(Ticket_Price_Cr) as Total_Ticket_price , Developer_Name from housing_data group by Developer_Name order by Total_Ticket_price DESC LIMIT 10;')
Data = mycursor.fetchall()
df_10 = pd.DataFrame(Data,columns=['Total_bookings','Total_Ticket_price','Developer_Name'])
df_10.to_csv('df_10',index = False)
```
